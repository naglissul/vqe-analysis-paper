```python
molecules = [
    ("H2", lambda d: f"H 0.0 0.0 0.0\nH 0.0 0.0 {d}"),
    ("LiH", lambda d: f"Li 0.0 0.0 0.0\nH 0.0 0.0 {d}"),
    ("BeH2", lambda d: f"Be 0.0 0.0 0.0\nH 0.0 0.0 {d}\nH 0.0 0.0 {-d}"),
]

distances=[0.1, 0.2, 0.5, 0.735, 1.0, 1.5, 2.0, 3.0, 4.0]
```

```python
import tequila as tq
import time

# initialize molecule (also works over .xyz files --> see next example)
def run_upccgsd(geomstring, order=1):
    # geomstring="H 0.0 0.0 0.0\nH 0.0 0.0 1.6"
    start = time.perf_counter()
    mol = tq.Molecule(geometry=geomstring)

    # get the qubit hamiltonian
    H = mol.make_hamiltonian()

    # get the ansatz (circuit)
    U = mol.make_upccgsd_ansatz(order=order)

    # define the expectation value
    E = tq.ExpectationValue(H=H, U=U)

    # minimize the expectation value
    result = tq.minimize(E)

    end = time.perf_counter()

    nit = None
    for line in str(result.scipy_result).splitlines():
        if "nit:" in line:
            nit = int(line.split(":")[1].strip())
            break

    return (result.energy, end - start, nit)

# geomstring="H 0.0 0.0 0.0\nH 0.0 0.0 1.6"
# (result, runtime, nit) = run_upccgsd(geomstring)
```

```python
energies = []
runtimes = []
nits = []
```

```python
for i in range(len(distances)):
    geomstring = molecules[0][1](distances[i])
    print(f"distance: {distances[i]}")
    print(geomstring)
    (result, runtime, nit) = run_upccgsd(geomstring)
    print(result, runtime, nit)
    energies.append(result)
    runtimes.append(runtime)
    nits.append(nit)
```
