```python
from lib.UpCCGSD_lib_conda import make_hamiltonian, make_upccgsd
```

```python
molecules = [
    ("H2", lambda d: f"H 0.0 0.0 0.0\nH 0.0 0.0 {d}"),
    ("LiH", lambda d: f"Li 0.0 0.0 0.0\nH 0.0 0.0 {d}"),
    ("BeH2", lambda d: f"Be 0.0 0.0 0.0\nH 0.0 0.0 {d}\nH 0.0 0.0 {-d}"),
]

distances=[0.1, 0.2, 0.5, 0.735, 1.0, 1.5, 2.0, 3.0, 4.0]
```

```python
# H2 hamiltonians
for distance in distances:
    for molecule, geometry in molecules:
        geom = geometry(distance)
        hamiltonian = make_hamiltonian(geom)
        print()
        print(f"molecule: {molecule}, distance: {distance}")
        print(hamiltonian)
```

```python
# ansatzes
ks = [1,2,3]

for distance in distances:
    for molecule, geometry in molecules:
        for k in ks:
            geom = geometry(distance)
            ansatz = make_upccgsd(geom, order=k)
            print()
            print(f"molecule: {molecule}, distance: {distance}, k: {k}")
            print(ansatz)
```
